**Project : 2 Guess the Number Game Python Project (computer)**

```python
from math import e
import random
def guessNumber(x):
  random_Number:int = random.randint(1,x)
  guessLimit:int = 3
  userInput = None
  print(f"Welcome to the Number Guessing Game! You have {guessLimit} attempts to guess the number between 1 and {x}.")

  while guessLimit >= 1:
    print('yaaaHA AYAAA H')
    try:
     userInput = input(f'Guess a number between 1 and {x} (or type "exit" to quit): ').strip().lower()
     if(userInput in ['exit','quit']):
        print("Thanks for playing! Goodbye!")
        return
     userGuess = int(userInput)
     if(userGuess < 1 or userGuess  > x):
      print(f"Please enter a number between 1 and {x}.")
      continue
     elif(userGuess < random_Number):
       guessLimit -= 1
       print(f'Your guess is too low. You have {guessLimit} guesses left.')
     elif(userGuess > random_Number):
       guessLimit -= 1
       print(f'Your guess is too high. You have {guessLimit} guesses left.')
     else:
      print(f'Congratulations! You guessed the right number: {userGuess}')
      return

    except ValueError:
      print('Invalid input. Please enter a valid integer')
      return

  print(f'Sorry, you have used all your guesses. The correct number was {random_Number}.')


guessNumber(3)
```
